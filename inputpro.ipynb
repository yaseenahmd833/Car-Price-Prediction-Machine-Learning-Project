```python
import numpy as np
import pandas as pd
import pickle
```

```python
ln = pickle.load(open('linear_model.sav','rb'))

sc = pickle.load(open("sc_model.sav","rb"))


le1 = pickle.load(open("le1_model.sav", "rb"))
# transmission1 = le1.transform([transmission])

le2 = pickle.load(open("le2_model.sav", "rb"))
# owner_type1 = le2.transform([owner_type])

ohe1 = pickle.load(open("ohe1_model.sav","rb"))
# ohe_brand = ohe1.transform([[brand]])
```

```python
engine_cc = float(input("Enter Engine CC : "))
mileage_kmpl = float(input("Enter Mileage kmpl : "))
horsepower = float(input("Enter Horsepower : "))
km_driven = float(input("Enter km_driven : "))
maintenance_cost = float(input("Enter maintenance_cost : "))
accident_count = int(input("Enter accident_count : "))
warranty_left_years = float(input("warranty_left_years : "))

fuel_type = input("Enter fuel type (Hybrid / Diesel / Electric / Petrol): ")
transmission = input("Enter transmission (Manual / Automatic): ")
owner_type = input("Enter owner type (First / Second / Third): ")
brand = input("Enter brand (Audi / BMW / Ford / Honda / Hyundai / Kia / Toyota): ")


```

```python
if fuel_type == "Hybrid":
    fuel_type = 1
elif fuel_type == "Diesel":
    fuel_type = 2
elif fuel_type == "Electric":
    fuel_type = 3
elif fuel_type == "Petrol":
    fuel_type = 4

fuel_type
```

```python
transmission_1 = le1.transform([transmission])[0]
```

```python
owner_type_1 = le2.transform([owner_type])[0]
```

```python
brand_ohe = ohe1.transform([[brand]])    # already numpy array
```

```python
# ---------- FINAL INPUT (ORDER MATTERS) ----------
numerical_part = np.array([[
    fuel_type,
    transmission_1,
    owner_type_1,
    engine_cc,
    mileage_kmpl,
    horsepower,
    km_driven,
    maintenance_cost,
    accident_count,
    warranty_left_years
]])

final_input = np.hstack((numerical_part, brand_ohe))
```

```python
final_scaled = sc.transform(final_input)
```

```python
# ---------- PREDICTION ----------
prediction = ln.predict(final_scaled)

print("Predicted Price:", prediction[0])
```
